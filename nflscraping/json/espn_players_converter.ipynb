```python
import json
import pandas as pd

with open('./espn_rosters.json') as rosters:
    data = json.load(rosters)
```

```python
urls = []
url_dict = {}

for url in data:
    urls.append(url['player_url'])

urls = list(set(urls))    



urls_dict = { i : urls[i] for i in range(0, len(urls)) }
urls_dict
```

```python
with open("espn_players_urls.json", "w") as outfile:
    json.dump(urls_dict, outfile, indent=3)


#print(json.dumps(df.to_dict('records')))
```
