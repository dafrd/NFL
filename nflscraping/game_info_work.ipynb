```python
# import libraries
import pandas as pd
import json
```

```python
# loading of data
file = open("espnscores.json")
results = json.load(file)
results[:3]
```

```python
df = pd.DataFrame(results)
df.head()
```

```python
df["test"] = df["idgame"].apply(lambda x : str(x).strip("[").strip("]"))
```

```python
df["gamescast"] = df["boxscore"].apply(lambda x : x.replace("boxscore","game"))
df["gamescast"].head()
```

```python
textfile = open("url_games.txt", "w")
for element in df["gamescast"]:
    textfile.write(element + "\n")
textfile.close()
```
